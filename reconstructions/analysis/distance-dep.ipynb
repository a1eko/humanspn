```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mx = np.linspace(0, 150)
hx = np.linspace(0, 250)
f = 1.7
```

mpas = 1 + 1 / (1 + np.exp((30.47 - mx) / 5.8))
hpas = 1 + 1 / (1 + np.exp((43.02 - hx) / 3.89))

```python
mpas = 1 + 1 / (1 + np.exp((37.08 - mx) / 4.21))
hpas = 1 + 1 / (1 + np.exp((43.02 - hx) / 3.89))
```

```python
mnaf = 0.1 + 0.9 / (1 + np.exp((mx - 40) / 10))
hnaf = 0.1 + 0.9 / (1 + np.exp((hx - 40*f) / (10*f)))
```

```python
mkaf = 1 + 0.5 / (1 + np.exp((mx - 120) / (-30)))
hkaf = 1 + 0.5 / (1 + np.exp((hx - 120*f) / (-30*f)))
```

```python
mkas = 1 + 9 * np.exp((mx - 0) / (-5))
hkas = 1 + 9 * np.exp((hx - 0*f) / (-5*f))
```

```python
mcat = 0 + 1 / (1 + np.exp((mx - 120) / (-30)))
hcat = 0 + 1 / (1 + np.exp((hx - 120*f) / (-30*f)))
```

```python
fig = plt.figure(figsize=(8,4))

ax = fig.add_subplot(1, 2, 1)
ax.set_title('Mouse')
ax.set_xlabel('Distance to soma ($\mu$m)')
ax.plot(mx, mpas, label='pas')
ax.plot(mx, mnaf, label='naf')
ax.plot(mx, mkaf, label='kaf')
ax.plot(mx, mkas, label='kas')
ax.plot(mx, mcat, label='cat')
ax.set_xlim(left=0)
ax.set_ylim(0, 3)

ax = fig.add_subplot(1, 2, 2)
ax.set_title('Human')
ax.set_xlabel('Distance to soma ($\mu$m)')
ax.plot(hx, hpas)
ax.plot(hx, hnaf)
ax.plot(hx, hkaf)
ax.plot(hx, hkas)
ax.plot(hx, hcat)
ax.set_xlim(left=0)
ax.set_ylim(0, 3)

fig.legend(loc='center right');
```
