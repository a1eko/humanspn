```python
hoc = 'Cell_9.hoc'
name = hoc.split('.')[0]

h.load_file("stdrun.hoc")
h.load_file(hoc)

cmd = 'h.%s("./morphology/", "animal_1_n30_04102017_cel1.swc")' % name
cell = eval(cmd)
```

```python
celsius = 35.0
v_init = -86.0
holding_current = 0.0009
stimulus = 0.4465
delay = 700.0
duration = 2000.0
simtime = 3000.0
```

```python
stim1  = h.IClamp(cell.soma[0](0.5))
stim1.amp = holding_current
stim1.delay = 0.0
stim1.dur = simtime

stim2 = h.IClamp(cell.soma[0](0.5))
stim2.amp = stimulus
stim2.delay = delay
stim2.dur = duration
```

```python
time = h.Vector()
time.record(h._ref_t)

volt = h.Vector()
volt.record(cell.soma[0](0.5)._ref_v);
```

```python
h.v_init = v_init
h.celsius = celsius
h.tstop = simtime
cvode = h.CVode()
cvode.active(1)
h.run();
```
